```python
nba_df_det = pd.read_csv("games_details.csv")
#nba_df_det = nba_df_det.groupby("GAME_ID").sum().reset_index()
nba_df_cen = nba_df_det[["GAME_ID","PTS","START_POSITION"]]
nba_cen = pd.merge(nba_df_cen,nba_df_game, on='GAME_ID')
nba_cen
```

```python
# There are some NaN values which renders that row unusable for our case.
nba_cen = nba_cen.groupby(["START_POSITION","SEASON"]).sum().reset_index()
nba_cen[nba_cen["SEASON"]==2003]
```

```python
nba_cen
```

```python
nba_cen_pivot = nba_cen.pivot(index = "SEASON",columns = "START_POSITION", values = "PTS")
```

```python
nba_cen_pivot.plot(kind="bar",figsize=(15,12))
plt.title('Point Distribution Between Positions')
plt.xlabel('Season')
plt.ylabel('Points')
plt.legend(["Center","Forward","Guard"])
```

```python
nba_cen_pivot_per = nba_cen_pivot
nba_cen_pivot_per[["C","F","G"]] = nba_cen_pivot_per[["C","F","G"]].div(nba_cen_pivot_per[["C","F","G"]].sum(axis=1),axis=0).multiply(100) 
fig, sub_p = plt.subplots(3,1, figsize=(30,80))
fig.tight_layout(pad=30.0)
#scatter(df.columns,df.loc[1994])
#sub_p[0,0].nba_cen_pivot_per["C"].plot(kind="bar")
#df1.plot(ax=axes[0])
#df2.plot(ax=axes[1])
fig.suptitle("Point Distribution Between Positions", fontsize=16)
nba_cen_pivot_per["C"].plot(ax=sub_p[0],figsize=(17,15))
sub_p[0].set_title("Center")
sub_p[0].set_ylabel("Seasons")
sub_p[0].set_ylabel("Points")
sub_p[0].set_xticks(np.arange(2003, 2021, 1))
nba_cen_pivot_per["F"].plot(ax=sub_p[1])
plt.xticks(np.arange(2003, 2021, 1))
sub_p[1].set_title("Forward")
sub_p[1].set_ylabel("Seasons")
sub_p[1].set_ylabel("Points")
sub_p[1].set_xticks(np.arange(2003, 2021, 1))
nba_cen_pivot_per["G"].plot(ax=sub_p[2])
plt.xticks(np.arange(2003, 2021, 1))
sub_p[2].set_xticks(np.arange(2003, 2021, 1))
sub_p[2].set_title("Guard")
sub_p[2].set_ylabel("Seasons")
sub_p[2].set_ylabel("Points")
```

```python
nba_df_pg = nba_df_det[["GAME_ID","START_POSITION","AST","PTS"]]
nba_df_game = nba_df[["GAME_ID","SEASON"]]
nba_pg = pd.merge(nba_df_pg,nba_df_game, on='GAME_ID')
nba_pg
```

```python
nba_pg = nba_pg.groupby(["START_POSITION","SEASON"]).sum().reset_index()
nba_pg = nba_pg.loc[nba_pg["START_POSITION"] == "G"]
nba_pg.set_index("SEASON",inplace = True)
```

```python
ax3 = nba_pg.plot.bar(y=['AST','PTS'],figsize = (15,10))
plt.title('Number of Assists and Points of guards')
plt.yticks(np.arange(0, max(nba_pg["PTS"]), 5000))
plt.xlabel('Season')
plt.ylabel('Amount')
plt.legend(["Assist","Points"])
```

As can be seen from the graph, for the first decade (2003-2010) points have lower average value and lower maximum value compared to the second decade (2011-2020). Approximately average of the second decade is the maximum points per season of the first decade. To justify this, dataset is divided into two as two decades and their statistical information are gathered. For the statistical data to be accurate, 2020 season data will be dropped as that seasons was halted because of Covid 19. Because of the lockout on 2011, that season will also be dropped.

Statistical data confirms the insight from the graph, maximum points recorded on the first decade is 81739.875 whereas the average of the points recorded on the second decade is 79130.125. When averages are compared,there is an approximately %5.6 increase and when maximums are compared the increase is greater. There is an approximately %8 percent increase between two maximum values.

When assist numbers are compared differences are much more smaller but this time first decade values are higher. There is a decrease around %1 for the averages and the maximums.

Results of these data are inline with the results of the other questions. On the second part, it was seen that the role of centers remained the same whereas forwards' role decreased and guards' role increased. Points scored by guards also confirmed this as a significant increase in points without much difference on the assists indicate a ball usage increase with the guards.
